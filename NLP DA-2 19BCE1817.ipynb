#### 1.	Create a text corpus with minimum 200 words (unique contents). Implement the following text processing                    
            •	Word segmentation

            •	Convert to Lowercase
            •	Stop words removal

            •	Part of speech tagger


```python
# Importing packages
import nltk
from nltk import word_tokenize
from nltk import sent_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk import PorterStemmer
from nltk import WordNetLemmatizer
```

```python
# Reading the content file and making it into one big paragraph string
with open('da2.txt', 'r') as file:
    data = file.read().replace('\n', '')
```

```python
# Word Segmentation

word_tokens=word_tokenize(data)
print(word_tokens)
print("\n"+"Number of word tokens = ",len(word_tokens))
```

```python
# Sentence Segmentation
sent_tokens=sent_tokenize(data)
print(sent_tokens)
print("\n"+"Number of sentences = ",len(sent_tokens))
```

```python
# Converting word tokens to lower case
word_tokens=[token.lower() for token in word_tokens]

print(word_tokens)
print("\n"+"Number of word tokens = ",len(word_tokens))
```

```python
# Stop Words Removal
stop = set(stopwords.words("english"))
stop_removed = [word_token for word_token in word_tokens if word_token not in stop]
print(stop_removed)
print("\n"+"Number of word tokens = ",len(stop_removed))
```

```python
def remove_punc(string):
    punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, "") 
    return string
 
final_tokens = [remove_punc(word) for word in stop_removed]
print(final_tokens)
print("\n"+"Number of tokens = ",len(final_tokens))
```

```python
# stemming using Porter Stemmer algorithm
ps = PorterStemmer()
for word in stop_removed:
    print(word+":"+ps.stem(word))
```

```python
# Lemmatization 
lemmatizer = WordNetLemmatizer()
for word in stop_removed:
    print(word+":"+lemmatizer.lemmatize(word))
```

```python
# POS Tagging
tagged = nltk.pos_tag(stop_removed)
print(tagged)
```
